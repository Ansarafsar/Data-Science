# Data Preprocessing

```python
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
```

```python
df1=df.copy()
```

```python
df1.drop(high_corr_cols, inplace=True, axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_encoded = df1.apply(le.fit_transform)
df_encoded
```

```python
df_encoded['y'].value_counts()
```

# Train Test Split Data

```python
x = df_encoded.drop('y',axis=1)  
y = df_encoded['y']              
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Function to evaluate the model

```python
def eval_model(y_test, y_pred):
    acc = accuracy_score(y_test, y_pred)
    print('Accuracy Score:', acc)
    cm = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix:\n', cm)
    print('Classification Report:\n', classification_report(y_test, y_pred))
    
    # Confusion Matrix Visualization
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", cbar=False,
                xticklabels=['No Purchase', 'Purchase'], yticklabels=['No Purchase', 'Purchase'])
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

# Model Training and Evaluation for Gini

```python
print("Training and Evaluating Decision Tree Classifier with Gini criterion")
dt_gini = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_split=10)
dt_gini.fit(x_train, y_train)
y_pred_gini = dt_gini.predict(x_test)
eval_model(y_test, y_pred_gini)
```

# Visualize the Decision Tree for Gini

```python
plt.figure(figsize=(20, 10))
plot_tree(dt_gini, feature_names=x.columns, class_names=["No Purchase", "Purchase"], filled=True)
plt.title("Decision Tree (Gini)")
plt.show()
```

# Model Training and Evaluation for Entropy


```python
print("Training and Evaluating Decision Tree Classifier with Entropy criterion")
dt_entropy = DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_split=10)
dt_entropy.fit(x_train, y_train)
y_pred_entropy = dt_entropy.predict(x_test)
eval_model(y_test, y_pred_entropy)
```

# Visualize the Decision Tree for Entropy

```python
plt.figure(figsize=(20, 10))
plot_tree(dt_entropy, feature_names=x.columns, class_names=["No Purchase", "Purchase"], filled=True)
plt.title("Decision Tree (Entropy)")
plt.show()
```

# Hyperparameter Tuning using GridSearchCV for Gini

```python
print("Hyperparameter Tuning for Gini criterion")
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(dt_gini, param_grid, cv=5)
grid_search.fit(x_train, y_train)
best_dt_gini = grid_search.best_estimator_
print("Best parameters for Gini found: ", grid_search.best_params_)
```

# Hyperparameter Tuning using GridSearchCV for Entropy

```python
print("Hyperparameter Tuning for Entropy criterion")
grid_search_entropy = GridSearchCV(dt_entropy, param_grid, cv=5)
grid_search_entropy.fit(x_train, y_train)
best_dt_entropy = grid_search_entropy.best_estimator_
print("Best parameters for Entropy found: ", grid_search_entropy.best_params_)
```

# Feature Importance Analysis for Gini

```python
feature_importances_gini = best_dt_gini.feature_importances_
feature_names = x.columns
sorted_indices_gini = np.argsort(feature_importances_gini)[::-1]

plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_indices_gini)), feature_importances_gini[sorted_indices_gini], align='center')
plt.yticks(range(len(sorted_indices_gini)), feature_names[sorted_indices_gini])
plt.title('Feature Importance for Gini')
plt.show()
```

# Feature Importance Analysis for Entropy

```python
feature_importances_entropy = best_dt_entropy.feature_importances_
sorted_indices_entropy = np.argsort(feature_importances_entropy)[::-1]

plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_indices_entropy)), feature_importances_entropy[sorted_indices_entropy], align='center')
plt.yticks(range(len(sorted_indices_entropy)), feature_names[sorted_indices_entropy])
plt.title('Feature Importance for Entropy')
plt.show()
```

# ROC Curve and AUC for Gini

```python
y_prob_gini = best_dt_gini.predict_proba(x_test)[:, 1]
fpr_gini, tpr_gini, _ = roc_curve(y_test, y_prob_gini)
roc_auc_gini = auc(fpr_gini, tpr_gini)

plt.figure(figsize=(8, 6))
plt.plot(fpr_gini, tpr_gini, color='blue', label='AUC (Gini) = %0.2f' % roc_auc_gini)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Gini')
plt.legend(loc='lower right')
plt.show()
```

# ROC Curve and AUC for Entropy

```python
y_prob_entropy = best_dt_entropy.predict_proba(x_test)[:, 1]
fpr_entropy, tpr_entropy, _ = roc_curve(y_test, y_prob_entropy)
roc_auc_entropy = auc(fpr_entropy, tpr_entropy)

plt.figure(figsize=(8, 6))
plt.plot(fpr_entropy, tpr_entropy, color='blue', label='AUC (Entropy) = %0.2f' % roc_auc_entropy)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Entropy')
plt.legend(loc='lower right')
plt.show()
```
